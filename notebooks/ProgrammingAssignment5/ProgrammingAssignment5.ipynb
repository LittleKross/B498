2) Write Python code to take the means and run the Mann Whitney U-test on the ENTRIESn_hourly column in the turnstile_weather dataframe.

you should show the following:
- i) the mean of entries with rain

- ii) the mean of entries without rain

- iii) the Mann-Whitney U-statistic and p-value comparing the number of entries with rain and the number of entries without rain

You should feel free to use scipy's Mann-Whitney implementation, and you might also find it useful to use numpy's mean function.

- iv) Is the distribution of the number of entries statistically different between rainy and non-rainy days?

Here are the functions' documentation:

http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mannwhitneyu.html

http://docs.scipy.org/doc/numpy/reference/generated/numpy.mean.html

See example below of Mann-Whitney for x1 and x2 created earlier. 

```python
# Answer Question #2 here

print('mean of the entries with rain, ' + str(rainData.mean()) + '\nmean of the entries without rain' + str(notRainData.mean()))

stat, pval = stats.mannwhitneyu(rainData, notRainData)
print('\nU-statistic: ' + str(stat) + '\nP-value: ' + str(pval))
if pval <= 0.05: # 95% confidence 
    print('\nReject the null hypothesis, there is a difference  ', pval)
else:
    print('\nAccept the null hypothesis, there is NO difference  , ', pval)
```
